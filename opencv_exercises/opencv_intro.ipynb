```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#read the image
img = cv2.imread('data/lena.png') #../Documents/AINLP/data/right.png
```

```python
#access image properties
print(f'The width is: {img.shape[1]}')
print(f'The height is {img.shape[0]}')
print(f'The shape is {img.shape}')
print(img.shape)
```

```python
#show image on the screen
cv2.imshow('Loaded image',img)
cv2.waitKey(0) #so that the image is on the screen for given amout of time - its in miliseconds
#Very nice!
```

```python
#write image on disk
cv2.imwrite('data/lena.jpg', img) # you did change extension png --> jpg
```

```python
#create empty image
empty = np.zeros((200,200), dtype=np.uint8) #np.zeros here we us enumpy function:)
print(empty)
cv2.imwrite('data/empty.jpg', empty) #gets written in the catalogue we are in. its black image square
```

```python
b,g,r = img[0,0] #values of the first pixel, top left
print(b,g,r) #opencv read the image as BGR not RGB
```

```python
plt.imshow(img) #myplotlib - how shows images:)
```

```python
img[:150,:150] = (0,0,0) #this makes a black quare on the image # (255,0,0) makes the square blue, (255,0,255) makes it magenta
cv2.imshow('Loaded image',img)
cv2.waitKey(0) #wait and stay on the screeen
```

```python
img[:150,:150] = (255,255,0) #is cyan
cv2.imshow('Loaded image',img)
cv2.waitKey(0) #wait and stay on the screeen
```
